```python
import pandas as pd
```

```python
crime_df = pd.read_csv('../data/On_Street_Crime_In_Camden.csv')

print(crime_df.info())

crime_df.head()
```

```python
# crime_df = crime_df[:1000]

crime_df['Location'].nunique()
```

```python
crime_df['Coordinate'] = crime_df.apply(lambda x: (x['Longitude'], x['Latitude']), axis=1)

crime_df.head()
```

```python
crime_freq = crime_df.groupby('Coordinate').size()
```

```python
crime_freq.count()
```

```python
print(crime_freq.min())
print(crime_freq.max())
```

```python
import json

with open('../data/template.geojson', 'r') as template:
    template = json.load(template)
    
    for index, value in crime_freq.iteritems():
        data = {
            'type': 'Feature',
            'properties': {
                'crimes': value
            },
            'geometry': {
                'type': 'Point',
                'coordinates': [index[0], index[1]]
            }
        }
        
        template['features'].append(data)

with open('../crimes.geojson', 'w') as output:
    json.dump(template, output)
```
